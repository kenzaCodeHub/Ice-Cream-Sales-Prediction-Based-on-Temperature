```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.linear_model import LinearRegression
```

```python
# Provided data
temperatures = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]  # in Celsius
ice_cream_sales = ['$215', '$325', '$185', '$332', '$406', '$522', '$412', '$614', '$544', '$421', '$445', '$408']# sales in dollars

# Convert ice cream sales to numpy array for calculations
ice_cream_sales_array = np.array([float(s[1:]) for s in ice_cream_sales])  # Remove the dollar sign and convert to float

# Calculate the linear regression line
slope, intercept = np.polyfit(temperatures, ice_cream_sales_array, 1)
regression_line = slope * np.array(temperatures) + intercept

# Plot the data points and linear regression line
plt.figure(figsize=(10, 6))
plt.scatter(temperatures, ice_cream_sales_array, color='b', label='Ice Cream Sales')
plt.plot(temperatures, regression_line, color='r', label='Linear (Ice Cream Sales)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales')  # Removed the dollar symbol from the y-axis label
plt.title('Ice Cream Sales')

# Formatting y-axis tick labels to include dollar symbol
formatter = ticker.FormatStrFormatter('$%1.0f')
plt.gca().yaxis.set_major_formatter(formatter)

plt.legend()
plt.grid(True)
plt.show()
```

```python
#Convert ice cream sales to integers
ice_cream_sales_int = [int(s[1:]) for s in ice_cream_sales]
```

```python
temperatures_np = np.array(temperatures).reshape(-1, 1)  # Reshape temperatures to a column vector
ice_cream_sales_np = np.array(ice_cream_sales_int).reshape(-1,1)  # Convert ice cream sales to a numpy array
```

```python
model = LinearRegression()
model.fit(temperatures_np, ice_cream_sales_np)
```

```python
# Evaluating the model quality
r_sq = model.score(temperatures_np,ice_cream_sales_np)
print('score of r square: ', r_sq)
```

```python
# examining the fitted model coefficents
print('slope:', model.coef_)
print('intercept:', model.intercept_)
```

```python
# Plot the data points
plt.scatter(temperatures_np, ice_cream_sales_np, color='blue', label='Data')

# Plot the linear regression line
plt.plot(temperatures_np, model.predict(temperatures_np), color='red', label='Linear Regression')

# Add labels and a legend
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales')
plt.title('Linear Regression: Ice Cream Sales vs Temperature')
plt.legend()

# Show the plot
plt.show()
```

```python
# Predict ice cream sales for a temperature of 15°C
temperature_15C = [[15]]  # Reshape the input to match the model's input shape
predicted_sales = model.predict(temperature_15C)

# Print the predicted ice cream sales
print("Predicted ice cream sales for a temperature of 15°C:", predicted_sales)
```
